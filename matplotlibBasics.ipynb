```python
x=[2,3,4,4,5,7,8,3]
y=[5,7,9,9,11,15,17,7]
x2=np.arange(0,10,0.5)

plt.savefig("myplot.png",dpi=300)
plt.figure(figsize=(7,5),dpi=100)#first set the size and then do plt.plot
plt.plot(x2[:5],x2[:5]**2,color="y",marker="*",markerfacecolor="red",markeredgecolor="black",markersize=10,label="X^2")
plt.plot(x2[4:],x2[4:]**2,color="black",label="X^2",linestyle="--")

plt.plot(x,y,label='2x+1',color="green",linewidth=2,marker='X',markeredgecolor="y",markersize=5,linestyle=":")
plt.title("Revising Matplotlib")
plt.legend()
plt.xlabel("Age")
plt.ylabel("number of people")
#plt.xticks([0,1,2,3,4,5])
#plt.yticks([0,1,2,3,10])
plt.arrow(3.9,2.9,0.1,0.1)#marks an arrow on the plot
plt.annotate("Change",(4,3))
plt.show()
```
